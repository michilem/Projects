```python
import matplotlib.pyplot as plt

values = [12, -21, -20, 19, 19, 19, 19, 18, 19, -21, 18, -23, 18, 16, 18, -22, 19, 19, 19, 18, 18] 
xs = range(len(values))

plt.axvline(0, color='r')
plt.axvline(8, color='r')
plt.axvline(14, color='r')
plt.plot(xs, values)
plt.xticks(range(0, len(values), 2))
plt.xlabel('Episode')
plt.ylabel('Cumulative reward')
plt.legend([
    'Goalie left',
    'Goalie right',
    'Goalie middle'],
    loc='center left'
)
plt.savefig('cumulative_reward.png')
plt.show()
```
